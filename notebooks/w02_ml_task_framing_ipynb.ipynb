## 1. My Lane as an ML Task

**Lane:** Refresh / Content Opportunity Scoring

**ML Task Type:** Ranking / Scoring

The objective is to rank content pages according to their likelihood of needing review. Instead of simply classifying pages as good or bad, the model produces a priority score so content reviewers can focus on the most important pages first.

## 2. Target or Proxy

The target is whether a content page is showing a declining trend and should be considered for review.

For the starter dataset, I will use `trend_direction == "down"` as a proxy label. This label comes from the current observation window, so it is not a perfect future outcome. It is useful for initial experimentation, but a stronger capstone would use past features to predict decline in a later time window.

## 3. Success Metric

The primary success metric is **Precision@50**. Since the goal is to help reviewers prioritize a limited number of pages, Precision@50 measures how many of the top 50 recommended pages are actually good review candidates. A higher Precision@50 means the ranking is more useful for real content review decisions.

## 5. Why ML Beats a Fixed Rule Here

A fixed rule typically considers only one or two conditions, such as low CTR or high content age. However, content performance depends on multiple interacting factors including impressions, search position, engagement, freshness, and traffic trends. Machine learning can learn these combined patterns and rank pages more effectively than a single manually defined rule, while still supporting human review and decision-making.
